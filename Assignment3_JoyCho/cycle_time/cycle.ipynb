# Part 3.2 Comparing and Gathering Data

```python
import ctypes
import time
import os
```

```python
_libCycle = ctypes.CDLL('./cycleLib.so')

# initialize the cycle counter
print(_libCycle.initialize_cntrs(1,1))

for i in range(5):
    print(f"\ntrial {i+1}")
    # get "before" time and cycle count
    before_time = time.time()
    before_cyclecnt = _libCycle.get_cycle_count()

    # run recur_fibo
    os.system("python3 recur_fibo.py")

    # get "after" time and cycle count
    after_time = time.time()
    after_cyclecnt = _libCycle.get_cycle_count()

    # cycle count and amount of time used
    duration = after_time - before_time
    cycle_count = after_cyclecnt - before_cyclecnt

    print(f"amount of time spent: {duration:.5f}")
    print(f"cycle count: {cycle_count}")
```
